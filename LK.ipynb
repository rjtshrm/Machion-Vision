Optical flow between two images with short displacement ( For large displacement create image pyramids)

```
O = min_sum_for_all_pixels(I(x + t) - H(x))²
I(x + t) = I(x) + t * I'(x) 
-> O = (I(x) + t * I'(x)  - H(x))²
minimize error o w.r.t to h
do / dt = 0
sum_((I(x) + t * I'(x)  - H(x)) * I'(x)) = 0
h = sum_all(I'(x) * (H(x) - I(x))) / sum_all(I'(x)²)
```
